# Steps in Exploratory Data Analysis (EDA)

**What is Exploratory Data Analysis (EDA)**

 Exploratory Data Analysis (EDA) is a data analysis technique that helps you understand your data by identifying patterns, anomalies, and relationships between variables. It's often the first step in any data analysis project.

## **1.Understand the data**

1.   **Load the Data** : Import the dataset using libraies like '**pandas**'
2.   **Preview the Data** : Use methods like '**head()**' ,**tail()** and  **sample()** to get a quick overview of the dataset.
3.   **Data dimensions** : Check the shape of the dataset useing **shape**
4.   **Data types**: Examine the data types of each column using **dtypes** or **info**




## **2.Data Cleaning**

1.  **Handle Missing Values** : Identify missing values using **isnull().sum()**
and decide on a strategy(eg. removal,imputation).
2.  **Remove Dulpilcate**: Check and remove duplicate rows using **duplicated()**
and **drop_duplicates()**.
3.   **Correct Data Types**: Convert Columns to appropriate data types(eg. datas,categprical)using **astype()**.



## **3.Descriptive Statistics**

1.   **Summary Statistics** : Use **describe()** for numerical columns and **value_counts()** for categorical columns to summarize the data.
2.   **Distribution Analysis** : Visalize the distribution of data using histograms


## **4.Data Visulization**

1. **Univeriate Analysis**: Plot individual features using histograms,box plots,and bar plots to understand thier distributions
2. **Bivariate Analysis**: Explore Relationships between two variables using scatter plots,correlation matrices and pair plots
3. **Multivariate Analysis** : Use heatmaps,pair plots and advanced visulaization to explore relatonship amoung multiple variable

## **5.Correlation Analysis**

1. **Correlation Matrix** : Calculate And visualize the correlation between numerical features using **corr()** and **heatmap()**.
2. **Feature Relationship**: Identify highly correlated features and consider features selection or dimensioality reduction (eg. PCA)

##**6.Outlier Detection:**



1.   **Visual Detection** : Use box plots and scatter plots to visually identify outliers
2.   **Statistical Detection** : Apply statistical methods (eg, Z-score,IQR) to detect outliers




## **7.Feature Engineering**



1.  **Create New Features** : Derive new features from existing data (eg, ratios,polynomial features).
2.   **Transform Features** : Apply Transformation (Eg, logarithmic,square root)
3.   **Encodeing Categorical Variables** : Convert categorical variable into numerical using methods like one-hot encoding (**pd.get_dummies()**) or label encoding





## **8.Dimensionality Reduction :**

1.   **PCA (Principal Component Analysis)**: Reduce the number of features while retaining variance using PCA
2.   **Feature Selection** : Select important features based on statistical tests, model importance , or correlation analysis



## **9.Data Scaling:**



1.   **Standardization** : Scale features to have zero mean and unit variance using **StandardScaler()**
2.   **Normalization** : Scale features to a range (eg, 0 to 1) using **MinMaxScaler()**.



## **10.Initial Modeling and Hypothesis Testing :**



1.   **Baseline Models**: Train simple models (e.g., linear regression) to establish a performance baseline.
2.  **Test Hypotheses**: Formulate and test hypotheses about the relationships between features and the target variable.




## **11. Documentation and Reporting:**



1.   **Summarize Findings**: Document key insights from the EDA process.
2.  **Visualization of Results**: Create visualizations to communicate findings effectively.

3.  **Prepare Data for Modeling**: Ensure that the dataset is clean, well- structured, and ready for modeling.

## **Step 7.Feature Engineering**

```python
# Create a new feature 'family_size'
df['family_size'] = df['sibsp'] + df['parch'] + 1

# Log transformation of 'fare' to reduce skewness
df['log_fare'] = np.log1p(df['fare'])        # Calculating the natural logarithm

# One-hot encode categorical variables (e.g., 'embarked')
df = pd.get_dummies(df, columns=['embarked'], drop_first=True)
```

## **Step 8.Dimensionality Reduction :**

**Principle Component Analysis (PCA)**

Dimensionality reduction technique used in machine learning

Dimensionality reduction is achieved by preserving the most important patterns or relationships between the variables with out any prior knowledge of the target variables.

**Application of PCA**
1.   Data Visualization- 6 dimension into 2 or 3 dimensions
2.   Feature Selection - Most variables in a dataset  can be identified
3.   Data Compression - Reduce size without lossing information


**Unsupervised Learning**

Orthogonal transformation that convert a set of correlated variables to a set uncorrelated variable





```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the features before PCA
features = ['age', 'fare', 'family_size']
X = StandardScaler().fit_transform(df[features])

# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)

# Add PCA results to the DataFrame
df['pca1'] = pca_result[:, 0]
df['pca2'] = pca_result[:, 1]

# Visualize PCA results
sns.scatterplot(x='pca1', y='pca2', hue='survived', data=df)
plt.title('PCA of Titanic Data')
plt.show()
```
